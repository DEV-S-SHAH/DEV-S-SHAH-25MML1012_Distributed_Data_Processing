## a) Theory

### Question:
Explain the concept of distributed data processing. How does Apache Spark improve performance compared to traditional data processing systems? Include the role of in-memory computation.

### Answer:

**Distributed Data Processing:**
- Distributed data processing divides large datasets across multiple machines (nodes) in a cluster
- Each node processes its portion of data in parallel
- Results are combined to produce the final output
- Enables processing of massive datasets that cannot fit on a single machine

**Apache Spark Performance Improvements:**
1. **In-Memory Computation**: Spark stores intermediate results in RAM instead of writing to disk, reducing I/O operations significantly
2. **DAG Optimization**: Uses Directed Acyclic Graph (DAG) execution engine to optimize task execution
3. **Lazy Evaluation**: Transforms are executed only when an action is called, allowing optimization of the entire pipeline
4. **Speed**: 100x faster than Hadoop MapReduce for in-memory operations, 10x faster for disk operations

**Role of In-Memory Computation:**
- Data is cached in memory across iterations
- Eliminates repetitive disk I/O operations
- Particularly beneficial for iterative algorithms (machine learning, graph processing)
- Traditional systems like MapReduce write intermediate results to disk after each operation, causing bottlenecks

## b) Practical / Coding

### Question:
Write a PySpark program to:
- Create a DataFrame with at least 5 records (columns: ID, Name, Age, Salary)
- Display the schema
- Show all records
- Filter employees with salary greater than 30,000

### Solution:

```python
# Import necessary libraries
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
# Create Spark Session
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

print("Spark Session Created Successfully!")
```

```python
# Define schema for the DataFrame
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

schema = StructType([
    StructField("ID", IntegerType(), True),
    StructField("Name", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("Salary", IntegerType(), True)
])
```

```python
# Create sample data with 5 employee records
data = [
    (1, "John Smith", 28, 25000),
    (2, "Sarah Johnson", 32, 45000),
    (3, "Mike Williams", 25, 28000),
    (4, "Emily Brown", 35, 55000),
    (5, "David Miller", 30, 38000)
]
```

```python
# Create DataFrame using schema and data
df = spark.createDataFrame(data, schema)
print("DataFrame created successfully!")
```

```python
# Display the schema of the DataFrame
print("\n=== DataFrame Schema ===")
df.printSchema()
```

```python
# Show all records in the DataFrame
print("\n=== All Employee Records ===")
df.show()
```

```python
# Filter employees with salary greater than 30,000
print("\n=== Employees with Salary > 30,000 ===")
filtered_df = df.filter(df.Salary > 30000)
filtered_df.show()
```

```python
# Alternative method using SQL-style filtering
print("\n=== Alternative Filtering Method ===")
filtered_df_alt = df.filter("Salary > 30000")
filtered_df_alt.show()
```

```python
# Display count of filtered records
print(f"\nTotal employees with salary > 30,000: {filtered_df.count()}")
```

```python
# Stop the Spark Session
spark.stop()
print("\nSpark Session stopped.")
```
